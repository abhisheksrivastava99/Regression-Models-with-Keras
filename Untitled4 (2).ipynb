Importing the important libraries

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

Downloading and cleaning the data

```python
concrete_data = pd.read_csv('Downloads/concrete_data.csv')
```

```python
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

splitting the data in target and predictors

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength']
```

```python
n_cols = predictors.shape[1] # number of predictors
n_cols
```

```python
predictors.head()
```

Part B-Normalizing predictors

```python
predictors_norm = (predictors- np.mean(predictors))/np.std(predictors)
predictors_norm.head()
```

Definig our baseline model

```python
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Splitting data randomly

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

Building the model

```python
model= baseline_model()
```

Training the model on 50 epochs

```python
model.fit(X_train, y_train, epochs=50, verbose=2)
```

Evaluating our model 

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

finding mean square error of our model

```python
mean_square_error = mean_squared_error(y_test, y_pred)
print(mean_square_error , np.mean(mean_square_error), np.std(mean_square_error))
```

Repeating step 1 to 3 and making a list of 50 mean squared errors

```python
mse=[]
for i in range(50):
    model = baseline_model()
    model.fit(X_train, y_train, epochs=50,verbose = 2)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
```
